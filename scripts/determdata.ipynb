```python
import os
import shutil
```

```python
source_dir = "/home/zhaob/Desktop/semantic-segmentation-pytorch/ethanCutUpDETdata/rainy_day/"
dest_dir = "/home/zhaob/Desktop/semantic-segmentation-pytorch/new_data"
```

```python
seeds = sorted([d for d in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, d))])

os.makedirs(dest_dir, exist_ok=True)

print("Found seeds:", seeds)

for idx, seed in enumerate(seeds, start = 1):
    seed_path = os.path.join(source_dir, seed)

    rgb_path = os.path.join(seed_path, "rgb")
    seg_path = os.path.join(seed_path, "rgb_seg")

    new_rgb_name = f"rgb_rainy_day{idx}"
    new_seg_name = f"rgb_seg_rainy_day_{idx}"

    new_rgb_path = os.path.join(dest_dir, new_rgb_name)
    new_seg_path = os.path.join(dest_dir, new_seg_name)

    print(f"Checking: {rgb_path}, {seg_path}")

    if os.path.exists(rgb_path):
        print(f"Moving {rgb_path} -> {new_rgb_path}")
        shutil.move(rgb_path, new_rgb_path)
    else:
        print(f"Warning: {rgb_path} does not exist!")

    if os.path.exists(seg_path):
        print(f"Moving {seg_path} -> {new_seg_path}")
        shutil.move(seg_path, new_seg_path)
    else:
        print(f"Warning: {seg_path} does not exist!")
    
```
